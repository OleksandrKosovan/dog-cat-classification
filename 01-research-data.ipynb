```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt
```

### Research

```python
DATA_PATH = 'data/'
```

```python
os.listdir(DATA_PATH)
```

```python
num_cat = len(os.listdir(os.path.join(DATA_PATH, 'cat')))
num_dog = len(os.listdir(os.path.join(DATA_PATH, 'dog')))
print("Number of cat images: ", num_cat)
print("Number of dog images: ", num_dog)
```

```python
objects = ('Dog', 'Cat')
y_pos = np.arange(len(objects))
performance = [num_dog,num_cat]

plt.bar(y_pos, performance, align='center', alpha=0.5)
plt.xticks(y_pos, objects)
plt.ylabel('Number')
plt.title('Number of pets. Bar chart')

plt.show()
```
